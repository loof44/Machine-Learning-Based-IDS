```python
import numpy as np
import pandas as pd
import os
import re

csv_files = {
 'CSV/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv': 'Clean_data/ddos.csv',
 'CSV/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv': 'Clean_data/web_attacks.csv',
 'CSV/Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv': 'Clean_data/infiltration.csv',
 'CSV/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv': 'Clean_data/port_scan.csv',
 
}

column_name_regex = re.compile(r"\W", re.IGNORECASE)










##############################################################################################################
def removing_duplicates(csv_files):
    for key, value in csv_files.items():
        df = pd.read_csv(key)
        df.drop_duplicates(inplace=True)
        df.to_csv(value, index=False)
        print(f"File {value} created")
#removing duplicates
#df.drop_duplicates()
#renaming feature names
def renaming_features(csv_files):
    for key, value in csv_files.items():
        df = pd.read_csv(key)
        column_name_regex = re.compile(r"\W", re.IGNORECASE)
        df.columns = [column_name_regex.sub('_', c.lower()) for c in df.columns]
        df.to_csv(value, index=False)
        print(f"File {value} created")
#df.columns = df.columns.str.lstrip('_')
#removing null values and replacing infinity values
#df2['flow_bytes_s'].replace(np.inf, 'inf', inplace=True)
#df2.dropna(inplace=True)
def removing_null_values(csv_files):
    for key, value in csv_files.items():
        df = pd.read_csv(key)
        df.dropna(inplace=True)
        df.to_csv(value, index=False)
        print(f"File {value} created")

def replacing_inf_values(csv_files):
    for key, value in csv_files.items():
        df = pd.read_csv(key)
        df['flow_bytes_s'].replace(np.inf, 'inf', inplace=True)
        df.to_csv(value, index=False)
        print(f"File {value} created")

# script to convert and clean all csv files
for key, value in csv_files.items():
    df = pd.read_csv(key)
    df.drop_duplicates(inplace=True)
    column_name_regex = re.compile(r"\W", re.IGNORECASE)
    df.columns = [column_name_regex.sub('_', c.lower()) for c in df.columns]
    df.columns = df.columns.str.lstrip('_')
    df['flow_bytes_s'].replace(np.inf, 'inf', inplace=True)
    df.dropna(inplace=True)
    df.to_csv(value, index=False)
    print(f"File {value} created")
```
