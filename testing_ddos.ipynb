```python
# getting correlation
corr = df_all.corr()
```

```python
#plotting the correlations
plt.figure(figsize=(20, 20))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, linewidths=.5, cmap=sns.diverging_palette(620, 10, as_cmap=True))
```

```python
#function to get strong correlations
def get_strong_correlations(correlation_matrix, threshold):
    strong_correlations = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i, j]) > threshold:
                strong_correlations.append([correlation_matrix.columns[i], correlation_matrix.columns[j], round(correlation_matrix.iloc[i, j], 3)])
    return strong_correlations
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(pd.DataFrame(get_strong_correlations(corr, 0.8), columns=['Feature 1', 'Feature 2', 'Correlation']))
```
